# shakti


```python
path_relative_binary_episodes = "data/3_binary_episodes"
name_file = 'binary_episodes_shakti_132010032025.csv'
name_file_trnng = 'binary_episodes_trnng_shakti_132010032025.csv'
path_binary_episodes = os.path.join(base_dir,path_relative_binary_episodes,name_file)
path_binary_episodes_trnng = os.path.join(base_dir,path_relative_binary_episodes,name_file_trnng)
df_binary_episodes_shakti = pd.read_csv(path_binary_episodes,sep=';')
df_binary_episodes_trnng_shakti = pd.read_csv(path_binary_episodes_trnng,sep=';')
```

```python
print(f"""
      ///////////////////////// shakti df: 
      shape: {df_binary_episodes_shakti.shape} \n 
      fail: {df_binary_episodes_shakti['fail'].value_counts()} \n
      ///////////////////////// shakti df trnng: 
      shape: {df_binary_episodes_trnng_shakti.shape}\n 
      fail: {df_binary_episodes_trnng_shakti['fail'].value_counts()}
      % fail: {df_binary_episodes_trnng_shakti['fail'].value_counts()[1]/(df_binary_episodes_trnng_shakti['fail'].value_counts()[1]+df_binary_episodes_trnng_shakti['fail'].value_counts()[0])}
      """)
```

```python
def training_model_random_forest(df:pd.DataFrame):
    x_train,x_test,y_train,y_test = train_test_split(df.drop('fail',axis=1),df['fail'],test_size=0.8)
    rnd_clf = RandomForestClassifier(
        n_estimators=500,
        n_jobs=-1,
    )
    rnd_clf.fit(x_train,y_train)
    y_pred = rnd_clf.predict(x_test)
    print(f"""
    {rnd_clf.__class__.__name__}
    len y_test: {y_test.shape}, shape y_pred: {y_pred.shape}
    accuracy: {accuracy_score(y_test,y_pred)}
    presicion score: {precision_score(y_test,y_pred)}
    recall score: {recall_score(y_test,y_pred)}
    confusion matrix:\n 
         {confusion_matrix(y_test,y_pred)}
     """)
    return rnd_clf,x_train,x_test,y_train,y_test
```

```python
rnd_forest_model_trnng_kapil,x_train_kapil,x_test_kapil,y_train_kapil,y_test_kapil = training_model_random_forest(df_binary_episodes_trnng_kapil)
rnd_forest_model_trnng_shakti,x_train_shakti,x_test_shakti,y_train_shakti,y_test_shakti = training_model_random_forest(df_binary_episodes_trnng_shakti)
rnd_forest_model_trnng_nihal,x_train_nihal,x_test_nihal,y_train_nihal,y_test_nihal = training_model_random_forest(df_binary_episodes_trnng_nihal)
```

```python
def save_model(
    model: sklearn.ensemble._forest.RandomForestClassifier,
    colums: list,
    autor: str,
    trnng: bool = False, 
    ):
    str_ftime = config.get(autor,'strftime_data_file')
    name_file = f'random_forest_model_{autor}_{str_ftime}.plk' if not trnng else f'random_forest_model_trnng_{autor}_{str_ftime}.plk'
    path_model_saved = os.path.join(
    base_dir, 
    config.get(autor,'path_random_forest_models_dir'),
    name_file,
    )
    joblib.dump((model,colums),path_model_saved)
    print(f'saved: {path_model_saved}')
    return 
```

```python
save_model(
 rnd_forest_model_trnng_kapil,
 x_test_kapil.columns.to_list(),
 'kapil',
 True,
)

save_model(
    rnd_forest_model_trnng_shakti,
    x_test_shakti.columns.to_list(),
    'shakti',
    True
)

save_model(
    rnd_forest_model_trnng_nihal, 
    x_test_nihal.columns.to_list(), 
    'nihal',
    True
)
```

```python
def construct_probability_table(
    autor: str,
    trnng: bool = False, 
    ):
    str_ftime = config.get(autor,'strftime_data_file')
    name_file = f'random_forest_model_{autor}_{str_ftime}.plk' if not trnng else f'random_forest_model_trnng_{autor}_{str_ftime}.plk'
    path_model_saved = os.path.join(
    base_dir, 
    config.get(autor,'path_random_forest_models_dir'),
    name_file,
    )
    name_file_binary = f'binary_episodes_{autor}_{str_ftime}.csv' if not trnng else f'binary_episodes_trnng_{autor}_{str_ftime}.csv'
    path_binary_episodes = os.path.join(
        base_dir, 
        config.get(autor,'path_binary_episodes_dir'),
        name_file_binary
    )
    name_file_abstract_episodes = f'abstract_episodes_{autor}_{str_ftime}.csv' if not trnng else f'abstract_episodes_trnng_{autor}_{str_ftime}.csv'
    path_file_abstract_episodes= os.path.join(
        base_dir, 
        config.get(autor,'path_abstract_episodes_dir'),
        name_file_abstract_episodes
    )

    model,columns = joblib.load(path_model_saved)
    df_binary = pd.read_csv(path_binary_episodes,sep=';')
    df_abstract_episodes = pd.read_csv(path_file_abstract_episodes,sep=';')

    df_binary = df_binary[columns]
    arr_prob = model.predict_proba(df_binary)
    print(arr_prob)
    print(arr_prob.shape)
    print(arr_prob.shape[1])
    arr_prob_fail = arr_prob[:,1]

    df_abstract_episodes['prob_fault'] = arr_prob_fail
    
    path_to_abstractprobabilityepisodes = os.path.join(
      base_dir, 
      config.get(autor,'path_abstract_datasets'),
      name_file_abstract_episodes, 
    )    
    df_abstract_episodes.to_csv(path_to_abstractprobabilityepisodes,sep=';')

    return df_abstract_episodes
```

```python
df_abstract_episodes_nihal = construct_probability_table(
    'nihal',
     True,
)
df_abstract_episodes_shakti= construct_probability_table(
    'shakti',
    True,
)
df_abstract_espisodes_kapil = construct_probability_table(
    'kapil',
    True, 
)
```
