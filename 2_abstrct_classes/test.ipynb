```python
import os
from pathlib import Path 
import configparser 
import json
import pandas as pd

base_dir = Path(os.getcwd()).resolve().parent
print('base dir', base_dir)
config = configparser.ConfigParser()
config.read(os.path.join(base_dir,'config.ini'))
```

```python
abstract_states_dir = config.get('nihal','path_abstract_states_dir')
q_values_states_dir = config.get('nihal','path_q_values_tables_dir')
name_file = 'abstract_states_4_nihal_382025032025.json'
file_abstract_states = os.path.join(base_dir,abstract_states_dir,name_file)
print(file_abstract_states)
```

```python
fjson = open(file_abstract_states,'r')
dic_abstrstates = json.load(fjson)
```
